#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
ds_filtrar = ds[ds['level5'].isin(['Valdemorillo', 'Galapagar'])]

average_price_valdemorillo = ds_filtrar[ds_filtrar['level5'] == 'Valdemorillo']['price'].mean()
average_price_galapagar = ds_filtrar[ds_filtrar['level5'] == 'Galapagar']['price'].mean()

ds_filtrar = ds_filtrar.dropna(subset=['price', 'surface']) 
ds_filtrar['pps'] = ds_filtrar['price'] / ds_filtrar['surface']

print(f"The average price in Valdemorillo is: {average_price_valdemorillo:.2f}")
print(f"The average price in Galapagar is: {average_price_galapagar:.2f}")

print("\nFiltered data with the 'pps' column added:")
print(ds_filtrar[['level5', 'price', 'surface', 'pps']])
```

#### Exercise 11. Analyze the relation between the surface and the price of the houses (★★☆)

Hint: You can make a `scatter plot`, then write a conclusion about it.

```python


import matplotlib.pyplot as plt

ds1 = ds[ds['surface'] < 5000]

plt.figure(figsize=(10, 6))
ds1.plot.scatter(x='surface', y='price', title='Relación Superficie-Precio', color='blue', alpha=0.6)

correlation = ds1['surface'].corr(ds1['price'])
print(f"Correlation between Surface and Price: {correlation}")

plt.xlabel('Surface (square meters)')
plt.ylabel('Price')

plt.show()
```

The correlation of 0.6421 indicates a moderate to strong positive connection between Surface and Price. In general, larger properties tend to be more expensive, but the relationship isn’t perfect. The fact that the correlation is not closer to 1 suggests that other factors, such as location, property condition, and amenities, also play a role in determining price.

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python

subset_ds['normalized_price'] = subset_ds.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

plt.figure(figsize=(10, 6))

for city in subset_ds['level5'].unique():
    plt.hist(subset_ds[subset_ds['level5'] == city]['normalized_price'], bins=20, alpha=0.5, label=city)

plt.title('Normalized Price Distribution by City')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')

plt.legend(title='City')
```

Based on the information above we can see that Alcorcon has a moderate number of properties with lower average price per meter and less variation. Fuenlabrada has a slightly higher average price per meter. Getafe and Leganes have similar average price per meter and has more diverse range of property.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python

subset_ds.loc[:, 'pps'] = subset_ds['price'] / subset_ds['surface']


getafe = subset_ds[subset_ds['level5'] == 'Getafe']
alcorcon = subset_ds[subset_ds['level5'] == 'Alcorcon']


mean_pps_getafe = getafe['pps'].mean()
mean_pps_alcorcon = alcorcon['pps'].mean()


variance_pps_getafe = getafe['pps'].var()
variance_pps_alcorcon = alcorcon['pps'].var()


print("Mean Price Per Square Meter for Getafe: " + str(mean_pps_getafe))
print("Variance in Price Per Square Meter for Getafe: " + str(variance_pps_getafe))
print("Mean Price Per Square Meter for Alcorcón: " + str(mean_pps_alcorcon))
print("Variance in Price Per Square Meter for Alcorcón: " + str(variance_pps_alcorcon))
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python

fig, axs = plt.subplots(2, 2, figsize=(12, 10))  
locations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

for i, city in enumerate(locations):
    ax = axs[i // 2, i % 2]  
    city_data = subset_ds[subset_ds['level5'] == city]  

    ax.scatter(city_data['surface'], city_data['pps'], alpha=0.5, s=20)  
    ax.set_title(f'{city}')
    ax.set_xlabel('Surface (m²)')
    ax.set_ylabel('Price Per Square Meter (€/m²)')

plt.tight_layout()  
plt.show() 
```
