#### Exercise 15. Make a bar plot of the median of the prices and explain what you observe (you must use the subset obtained in Exercise 14) (★★★)

Print the bar of the median of the prices and write in the Markdown cell a brief analysis about the plot.

```python

import matplotlib.pyplot as plt

df = pd.read_csv('assets/real_estate.csv', sep=';')


df["price"] = pd.to_numeric(df["price"], errors="coerce")
df = df.dropna(subset=["price"])  

median_prices = df.groupby("level5")["price"].median()

plt.figure(figsize=(12, 7))
plt.plot(median_prices.index, median_prices.values, marker="o", linestyle="-", color="blue")

plt.title("Median House Prices by Population (level5)")
plt.xlabel("Population (level5)")
plt.ylabel("Median Price (€)")
plt.xticks(rotation=45, ha="right")  
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('assets/real_estate.csv', sep=';')

subset_df = df[df["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcon"])]

subset_df['normalized_price'] = subset_df.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)


plt.figure(figsize=(10, 6))

for city in ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcon']:
    
    city_data = subset_df[subset_df['level5'] == city]
    
   
    plt.hist(city_data['normalized_price'], bins=20, alpha=0.5, label=city)

plt.title('Normalized Price Distribution by City')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend(title='City')

plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd
df = pd.read_csv('assets/real_estate.csv', sep=';')

subset_df = df[df["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcon"])]


subset_df = subset_df[subset_df['surface'] > 0]


subset_df.loc[:, 'pps'] = subset_df['price'] / subset_df['surface']
getafe_df = subset_df[subset_df['level5'] == 'Getafe']
alcorcon_df = subset_df[subset_df['level5'] == 'Alcorcon']


mean_pps_getafe = getafe_df['pps'].mean()
mean_pps_alcorcon = alcorcon_df['pps'].mean()


variance_pps_getafe = getafe_df['pps'].var()
variance_pps_alcorcon = alcorcon_df['pps'].var()


print(f"Mean Price Per Square Meter for Getafe: {mean_pps_getafe}")
print(f"Variance in Price Per Square Meter for Getafe: {variance_pps_getafe}")
print(f"Mean Price Per Square Meter for Alcorcón: {mean_pps_alcorcon}")
print(f"Variance in Price Per Square Meter for Alcorcón: {variance_pps_alcorcon}")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('assets/real_estate.csv', sep=';')


subset_df = df[df["level5"].isin(["Torrejón de Ardoz", "Navalagamella", "Rivas Vaciamadrid", "San Sebastián de los Reyes"])]

subset_df = subset_df[subset_df['surface'] > 0]

subset_df['pps'] = subset_df['price'] / subset_df['surface']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
cities = ['Torrejón de Ardoz', 'Navalagamella', 'Rivas Vaciamadrid', 'San Sebastián de los Reyes']

for i, city in enumerate(cities):
    ax = axs[i // 2, i % 2]  
    city_data = subset_df[subset_df['level5'] == city]
    
    
    ax.scatter(city_data['surface'], city_data['pps'], alpha=0.5, s=20)
    ax.set_title(f'{city}')
    ax.set_xlabel('Surface (m²)')
    ax.set_ylabel('Price Per Square Meter')


plt.tight_layout()
plt.show()
```
